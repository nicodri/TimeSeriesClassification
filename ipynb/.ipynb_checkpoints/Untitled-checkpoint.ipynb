```lua
require 'hdf5'
require 'nn'
```

```lua
Plot = require 'itorch.Plot'
```

```lua
myFile = hdf5.open('../data/split_50_0_4.hdf5','r')
f = myFile:all()
myFile:close()
```

```lua
f
```

```lua
nelements = 0
for i = 1, 21 do
    nelements = nelements + f['train_'..i]:size(1)
end

train_inputs = torch.Tensor(nelements, 70)
train_outputs = torch.Tensor(nelements)
k = 1
for i = 1, 21 do
    ts = f['train_'..i]
    for j = 1, ts:size(1) do
        ts = f['train_'..i][j]
        tt = torch.Tensor(70,3)
        tt:copy(f['train_'..1][1]:narrow(2,2,3))
        tt:pow(2)
        train_inputs[k]:copy(tt:sum(2):pow(.5))
        train_outputs[k] = i
        k = k + 1
    end
end
```

```lua
model = nn.Sequential()
model:add(nn.View(70,1))
model:add(nn.TemporalConvolution(1,15,5))
model:add(nn.Tanh())
model:add(nn.TemporalMaxPooling(3))

model:add(nn.TemporalConvolution(15,15,3))
model:add(nn.Tanh())
model:add(nn.TemporalMaxPooling(2))

model:add(nn.View(150))
model:add(nn.Linear(150, 75))
model:add(nn.Tanh())
model:add(nn.Linear(75, 22))
model:add(nn.LogSoftMax())
```

```lua
function train_model(train_inputs, train_outputs, model, criterion, eta, nEpochs)
    -- Train the model with a SGD
    -- standard parameters are
    -- nEpochs = 1
    -- eta = 0.01

    -- To store the loss
    local loss = torch.zeros(nEpochs)
    local av_L = 0
    local f = 0
    local df_do
    local len = train_inputs:size(2)
    for i = 1, nEpochs do
        -- Display progess
        xlua.progress(i, nEpochs)

        -- timing the epoch
        local timer = torch.Timer()
        av_L = 0
        
        for ii = 1, train_inputs:size(1) do
            
            -- reset gradients
            model:zeroGradParameters()
            --gradParameters:zero()

            -- Forward pass (selection of inputs_batch in case the batch is not full, ie last batch)
            local pred = model:forward(train_inputs[ii])
            -- Average loss computation
            f = criterion:forward(pred, train_outputs[ii])
            print(train_outputs[ii])
            av_L = av_L + f
            
            -- Backward pass
            df_do = criterion:backward(pred, train_outputs[ii])
            model:backward(train_inputs[ii], df_do)
            model:updateParameters(eta)
            
        end

        loss[i] = av_L/train_inputs:size(1)
        print('Epoch '..i..': '..timer:time().real)
       	print('\n')
        print('Average Loss: '.. loss[i])
        print('\n')
        print('***************************************************')
       	
    end

    return loss
end
```

```lua
criterion = nn.ClassNLLCriterion()
```

```lua
loss = train_model(train_inputs, train_outputs, model, criterion, 0.1, 10)
```
